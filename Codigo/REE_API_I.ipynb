**API - Header y configuración Token para hacer peticiones**




```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
API_TOKEN = 'API_TOKEN'
headers = {
    'Host': 'api.esios.ree.es', #Dominio del servidor al que está haciendo la solicitud
    'x-api-key': API_TOKEN, #Método de la API para utentificarse
    'Accept': 'application/json' #Pide que la respuesta sea en un json
}
```

```python
#Comprobamos la conexión
url = 'https://api.esios.ree.es'
response = requests.get(url, headers=headers)
```

```python
response #200, éxito en la petición
```

```python
#Vamos a extraer una lista con todos los indicadores
#Planteamos dos partes de la URl, una que sea estándar y otra para acceder a la info de los indicadores en concreto.
url_page = 'https://api.esios.ree.es/'
indicators_position = 'indicators/'

url = url_page + indicators_position
content = requests.get(url, headers=headers) #Almacenamos la respuesta de la API

data=content.json() #Convierte la respuesta a un json
```

```python
#Estructura del json Claves: 'name', 'description', 'short_name', 'id.
data
```

```python
#Recorremos los datos del json para verlos de forma más clara, con el ID y el Nombre de cada Indicador.
for x in data['indicators']:
    print(x['id'], x['name'])
```

```python
#Convertimos a un df, creamos un lista donde se irá guardando la info de los indicadores al momento que lo va recorriendo el bucle y lo convertimos a un df.
lista_indicadores = []

for x in data['indicators']:
    lista_indicadores.append({
        "id": x['id'],
        "name": x['name']
    })

df = pd.DataFrame(lista_indicadores)
print(df.head())
```

```python

start_date = "2025-01-01T00:00:00Z" #Fecha Inicial
end_date = "2025-02-20T23:59:59Z"  #Fecha Final
loc = "Valencia"
url = f"{url_page}{indicators_position}{id_indicator}?start_date={start_date}&end_date={end_date}&location={loc}"

res = requests.get(url, headers=headers)

#Verificamos que la petición a tenido éxito
if res.status_code == 200:
    data = res.json()


    lista_valores = []
    for values in data["indicator"]["values"]:
        lista_valores.append({
            "datetime": values["datetime"],  #Fecha y Hora
            "value": values["value"]         #Valor
        })


    df = pd.DataFrame(lista_valores)
    df['datetime'] = pd.to_datetime(df['datetime'])

    df = df.sort_values(by='datetime') #Ordenamos las fechas por si viene en un orden que no es correcto

    print(df.head())
    df.to_csv("datos_1_2025.csv", index=False)
else:
    print(f"Error {res.status_code}: {res.text}")
```
