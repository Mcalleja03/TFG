```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# Configuración API
AEMET_API_URL = "https://opendata.aemet.es/opendata/api/valores/climatologicos/diarios/datos/fechaini/{fecha_ini}/fechafin/{fecha_fin}/estacion/{estacion}"
TOKEN = "TOKEN_MARIO"
estacion_id = "5973"

def obtener_datos_aemet(fecha_inicio, fecha_fin, estacion_id):
    datos_totales = []
    fecha_actual = datetime.strptime(fecha_inicio, "%Y-%m-%d")

    while fecha_actual < datetime.strptime(fecha_fin, "%Y-%m-%d"):
        fecha_siguiente = fecha_actual + timedelta(days=180)
        if fecha_siguiente > datetime.strptime(fecha_fin, "%Y-%m-%d"):
            fecha_siguiente = datetime.strptime(fecha_fin, "%Y-%m-%d")

        url = AEMET_API_URL.format(
            fecha_ini=fecha_actual.strftime("%Y-%m-%dT00:00:00UTC"),
            fecha_fin=fecha_siguiente.strftime("%Y-%m-%dT23:59:59UTC"),
            estacion=estacion_id
        )
        headers = {"api_key": TOKEN}


        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            json_response = response.json()
            url_datos = json_response.get("datos", None)

            if url_datos:
                response_datos = requests.get(url_datos)
                if response_datos.status_code == 200:
                    datos_totales.extend(response_datos.json())


        fecha_actual = fecha_siguiente + timedelta(days=1)

    return pd.DataFrame(datos_totales)


fecha_inicio = "2015-01-01"
fecha_fin = "2025-04-12"

df_temperaturas = obtener_datos_aemet(fecha_inicio, fecha_fin, estacion_id)

print(df_temperaturas.head())
```
