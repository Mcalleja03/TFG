```python
import requests
import pandas as pd
from datetime import datetime, timedelta
```

```python
def generar_fechas_mensuales(inicio_ao, fin_ao):
    fechas = []
    for ao in range(inicio_ao, fin_ao + 1):
        for mes in range(1, 13):  # De mes en mes
            inicio = datetime(ao, mes, 1)
            if mes == 12:
                fin = datetime(ao + 1, 1, 1) - timedelta(seconds=1)
            else:
                fin = datetime(ao, mes + 1, 1) - timedelta(seconds=1)
            fechas.append((inicio, fin))
    return fechas


dfs = []
for inicio, fin in generar_fechas_mensuales(2015, 2025):
    start_str = inicio.strftime('%Y-%m-%dT%H:%M')
    end_str = fin.strftime('%Y-%m-%dT%H:%M')
    day = f'?start_date={start_str}&end_date={end_str}'
    url = URL_BASE + ENDPOINT + INDICATOR + day
    print(f'Haciendo peticin: {url}')
    res = requests.get(url, headers=headers)
    if res.status_code == 200:
        data = res.json()
        if 'indicator' in data and 'values' in data['indicator']:
            df = pd.DataFrame(data['indicator']['values'])
            if not df.empty:
                df = df[['datetime_utc', 'geo_id', 'geo_name', 'value']]
                dfs.append(df)
    else:
        print(f'Error {res.status_code} al hacer la peticin para: {start_str} a {end_str}')

if dfs:
    df_total = pd.concat(dfs, ignore_index=True)
    df_total.to_excel('datos_indicador_1293_mensual.xlsx', index=False)
    print("Archivo Excel guardado como 'datos_indicador_1293_mensual.xlsx'")
else:
    print("No se obtuvieron datos.")
```

```python
import os
import pandas as pd

ruta_carpeta = '/content/'

todos_los_dfs = []

for ao in range(2015, 2026):
    for mes in range(1, 13):
        nombre_archivo = f"{ao}_{mes}.xlsx"
        ruta_archivo = os.path.join(ruta_carpeta, nombre_archivo)

        if os.path.exists(ruta_archivo):
            print(f"Leyendo {nombre_archivo}...")
            df = pd.read_excel(ruta_archivo)
            df["Ao"] = ao
            df["Mes"] = mes
            todos_los_dfs.append(df)

df_unido = pd.concat(todos_los_dfs, ignore_index=True)

ruta_salida = os.path.join(ruta_carpeta, 'unificado.xlsx')
df_unido.to_excel(ruta_salida, index=False)

print(f"Excel combinado guardado en: {ruta_salida}")
```
